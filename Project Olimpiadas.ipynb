```python
import pandas as pd
```

```python
dados = pd.read_html('/content/Quadro de medalhas dos Jogos Olímpicos – Wikipédia, a enciclopédia livre.html')
dados[0]
```

```python
dados
```

```python
indices_para_remover = [10,9,8,7,6,5,1]

for index in sorted(indices_para_remover, reverse=True):
    dados.pop(index)

print(dados)
```

```python
quadro_de_medalhas = dados[0]
quadro_de_medalhas
```

```python
quadro_de_medalhas.columns
```

```python
quadro_de_medalhas = quadro_de_medalhas.rename(columns={"Unnamed: 1": "Verão", "Unnamed: 2": "Ouro(V)", "Unnamed: 3": "Prata(V)", "Unnamed: 4": "Bronze(V)", "Unnamed: 5": "Total(V)",'Unnamed: 6': 'Inverno', 'Unnamed: 7':'Ouro(I)', 'Unnamed: 8':'Prata(I)', 'Unnamed: 9':'Bronze(I)',
       'Unnamed: 10':'Total(I)', 'T':'Somatória', 'Unnamed: 12':'Ouro(S)', 'Unnamed: 13':'Prata(S)', 'Unnamed: 14':'Bronze(S)',
       'Unnamed: 15':'Total(S)'})
```

```python
quadro_de_medalhas
```

```python
quadro_de_medalhas['País'] = quadro_de_medalhas['País'].str.replace('\[.*?\]', '', regex=True)
```

```python
quadro_de_medalhas
```

```python
quadro_de_medalhas.info()
```

```python
dados[0] = quadro_de_medalhas
dados[0]
```

```python
CONs_sem_medalhas = dados[1]
CONs_sem_medalhas = CONs_sem_medalhas.rename(columns={"Unnamed: 1": "Verão", "Unnamed: 2": "Inverno", 'T':'Somatória'})
dados[1] = CONs_sem_medalhas
dados[1]
```

```python
quadro_verao = dados[2]
quadro_verao = quadro_verao.drop('No.', axis=1)
dados[2] = quadro_verao
dados[2]
```

```python
quadro_inverno = dados[3]
quadro_inverno = quadro_inverno.drop('No.', axis=1)
dados[3] = quadro_inverno
dados[3]
```

```python
dados
```

```python
dados[0].to_csv('Quadro de Medalhas.csv', index=False, sep=';')
dados[1].to_csv('CONs Sem Medalhas.csv', index=False, sep=';')
dados[2].to_csv('Quadro Verão.csv', index=False, sep=';')
dados[3].to_csv('Quadro Inverno.csv', index=False, sep=';')
```
